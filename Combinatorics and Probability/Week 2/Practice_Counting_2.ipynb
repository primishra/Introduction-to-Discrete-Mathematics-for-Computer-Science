```
## Number of non-negative integers with at most 4 digits are increasing.
```

```
from itertools import product

count = 0

for number in product(range(10), repeat = 4):
  if number[0] < number[1] and number[1] < number[2] and number[2] < number[3]:
    count += 1
    print(number)

print(count)
```
